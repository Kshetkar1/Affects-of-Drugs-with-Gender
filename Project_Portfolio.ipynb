```python
#Import packages
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
data_url = 'https://github.com/moggirain/Hospital_readmission_prediction/blob/master/diabetic_data.csv?raw=true'
```

```python
data_df = pd.read_csv(data_url)
data_df
```

```python
data_df_race = data_df["race"] == 'Caucasian'
data_df[data_df_race]
```

```python
data_df.columns
```

```python
# from os import truncate



# select the columns you want to include in the filtered data frame
filtered_df = data_df[['diabetesMed', 'gender']]

# group the filtered data frame by race and gender and count the number of rows in each group
grouped_df = filtered_df.groupby(['diabetesMed', 'gender']).size().reset_index(name='count')

# pivot the grouped data frame so that the gender is in columns and the counts are in rows
pivoted_df = grouped_df.pivot(index='diabetesMed', columns='gender', values='count')

# create a bar plot of the pivoted data frame
pivoted_df.plot(kind='bar', stacked= False, logy= True)

# set the x and y axis labels
plt.xlabel('diabetesMed')
plt.ylabel('Count (log scale)')

# set the title of the plot
plt.title('Number of males and females and others within each race')

# show the plot
plt.show()
```
